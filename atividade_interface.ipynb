```python
#importaÃ§Ã£o dos dados
#link https://www.kaggle.com/datasets/harrimansaragih/dummy-data-for-classification
import pandas as pd
url='https://drive.google.com/file/d/15arnNph9O3wcte5zLqUKlmt_hMVGd6xa/view?usp=share_link'
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
df = pd.read_csv(url,sep=',')
```

```python
new_df = df.dropna(axis=0)
```

```python
new_df['Region'] = new_df['Region'].replace({"Urban": 1, "Rural":0})
```

##Identificar os tipos de Dados do Dataset

```python
# Iniciar a configuraÃ§Ã£o
# Depois aperte ENTER para continuar.
from pycaret.classification import *
s = setup(new_df, target = 'Downloaded', session_id = 123)
```

##treinado com 14 modelos diferentes, A tabela mostra as mÃ©tricas do Maior para o Menor

```python
best = compare_models(n_select = 15)
compare_model_results = pull()
```

##Para criar a Interface foi usado a biblioteca Gradio

https://towardsdatascience.com/supercharge-your-machine-learning-experiments-with-pycaret-and-gradio-5932c61f80d9

```python
#PrediÃ§Ã£o
def predict(model, age, spending, region):
  
    df = pd.DataFrame.from_dict({'Age': [age], 'Spending': [spending], 'Region': [region]})
    
    model_index = list(compare_model_results['Model']).index(model)
    model = best[model_index]
    pred = predict_model(model, df, raw_score=True)
    return {'JÃ¡ fez Download Netflix': pred['Score_Yes'][0].astype('float64'), 'NÃ£o Fez o Download Netflix': pred['Score_No'][0].astype('float64')}
```

```python
import gradio as gr

model = gr.inputs.Dropdown(list(compare_model_results['Model']), label="Modelo")
age = gr.inputs.Slider(minimum=1, maximum=50, default=new_df['Age'].mean(), label = 'Idade')
spending = gr.inputs.Slider(minimum=1, maximum=10000, default=new_df['Spending'].mean(), label = 'Gastos')
region = gr.inputs.Slider(minimum=0, maximum=1, default=new_df['Region'].mean(), label = 'RegiÃ£o')

gr.Interface(predict, [model,age,spending,region], "label", live=True).launch(share= True)
```
